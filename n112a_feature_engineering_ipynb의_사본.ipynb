## 1. Feature Engineering

- `KT&G` 기업 데이터를 계속 사용합니다.

- 각 분기별 영업이익률을 직접 계산해보세요. 

- 18/19년도에 대해서 Price to Dream Ratio (PDR)을 계산하세요. 단 여기서 점유율은 일반담배 기준이며 18년:62%, 19년 63.5% 로 계산합니다. 시장의 규모는 18년 15조, 19년 16조 라고 가정합니다. 

참조 : 
<http://it.chosun.com/site/data/html_dir/2020/02/13/2020021302434.html>


```python
# KT&G 기업데이터 불러오기, 숫자는 억 단위
import pandas as pd

df = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/kt%26g/kt%26g.csv')

df
```

```python
# 각 분기별 영업이익률을 직접 계산해보세요.

df.dtypes
```

```python
# 계산 전 모두 숫자로 바꿔준다.
# 숫자로 바꿔주는 함수 선언, replce 쓰지 않음
def to_int(is_string):
    return int(''.join(is_string.split(',')))
```

```python
# 영업이익율에 필요한 영업이익, 매출액 컬럼 숫자로 변경
df['매출액_int'] = df['매출액'].apply(to_int)
df['영업이익_int'] = df['영업이익'].apply(to_int)
df.dtypes
```

```python
# 영업이익/매출액 X 100 = 영업이익률
# round 함수로 소숫점 자리 2째자리까지 표현
#################1번 답##################
df['영업이익직접'] = df['영업이익_int'] / df['매출액_int'] * 100
df['영업이익직접'] = round(df['영업이익직접'], 2)
df
```

```python
#18/19년도에 대해서 Price to Dream Ratio (PDR)을 계산하세요. 단 여기서 점유율은 일반담배 기준이며 18년:62%, 19년 63.5% 로 계산합니다. 
# 시장의 규모는 18년 15조, 19년 16조 라고 가정합니다.
# 시가총액을 TAM에 해당 기업의 시장점유율을 곱한 값으로 나눠 PDR을 산출한다. 
# 예를 들어 시가총액 5조원 규모의 A업체가 100조원의 시장 규모를 가진 산업 내에서 10%의 시장점유율을 갖고 있다면 PDR은 0.5다.

# 2018년 : 13.8조(시총)/15조*0.62
# 2019년 : 12조/16조*0.635
```

```python
#2018년  138,000 / (150,000*0.62)
#2019년  120,000/160,000*0.635

#변수 각 년도별 시가총액
mc_2018 = 138000
mc_2019 = 120000
#TAM 변수
tam_2018 = 150000
tam_2019 = 160000
# 시장점유율 변수
ms_2018 = 0.62
ms_2019 = 0.635

```

```python
pdr_dict = {'연도':['2018','2019']
            ,'시가총액':[mc_2018, mc_2019]
            ,'TAM':[tam_2018,tam_2019]
            ,'시장점유율':[ms_2018,ms_2019]
            }
```

```python
#데이터프레임만들기
df_pdr = pd.DataFrame(pdr_dict)
df_pdr.head()
df_pdr['PDR'] = df_pdr['시가총액']/(df_pdr['TAM']*df_pdr['시장점유율'])
#######################1.2번답#######################
# pdr 추가해서 dataframe 출력
df_pdr.head()
```

### Feature Engineering

Relative Perfomance 라는 새로운 feature를 계산하세요.

이는 최근 1년치 매출액의 평균값을 기준으로 

- 10% 이상 -> `S`
- 5% 이상 -> `A`
- -5 ~ 5% -> `B`
- -5%이하 -> `C`
- 10%이하 -> `D`

라는 값을 갖는 feature입니다.

20년도 2분기에 해당하는 결과값은 `A`가 나와야합니다.

각각에 해당하는 등급이 나오기 위해서 필요한 매출액을 추가로 서술하세요.

```python
#우선 평균 먼저 구한다.

df_sale = df_plus[['분기','매출액']]
df_sale
# 최근 1녀간 평균 매기는데 필요없는 0번째 인덱스 삭제
df_sale_drop = df_sale.drop(index=0, axis=1)
# 숫자로 바꿔준다.
df_sale_drop['매출액'] = df_sale_drop['매출액'].apply(to_int)
```

```python
# mean()을 써서 평균 구한다.
sale_mean = df_sale_drop.mean()
print(sale_mean)
```
