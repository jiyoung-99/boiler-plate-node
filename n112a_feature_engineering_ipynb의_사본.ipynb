```python
# ê³„ì‚° ì „ ëª¨ë‘ ìˆ«ìë¡œ ë°”ê¿”ì¤€ë‹¤.
# ìˆ«ìë¡œ ë°”ê¿”ì£¼ëŠ” í•¨ìˆ˜ ì„ ì–¸, replce ì“°ì§€ ì•ŠìŒ
def to_int(is_string):
    return int(''.join(is_string.split(',')))
```

```python
# ì˜ì—…ì´ìµìœ¨ì— í•„ìš”í•œ ì˜ì—…ì´ìµ, ë§¤ì¶œì•¡ ì»¬ëŸ¼ ìˆ«ìë¡œ ë³€ê²½
df['ë§¤ì¶œì•¡_int'] = df['ë§¤ì¶œì•¡'].apply(to_int)
df['ì˜ì—…ì´ìµ_int'] = df['ì˜ì—…ì´ìµ'].apply(to_int)
df.dtypes
```

```python
# ì˜ì—…ì´ìµ/ë§¤ì¶œì•¡ X 100 = ì˜ì—…ì´ìµë¥ 
# round í•¨ìˆ˜ë¡œ ì†Œìˆ«ì  ìë¦¬ 2ì§¸ìë¦¬ê¹Œì§€ í‘œí˜„
#################1ë²ˆ ë‹µ##################
df['ì˜ì—…ì´ìµì§ì ‘'] = df['ì˜ì—…ì´ìµ_int'] / df['ë§¤ì¶œì•¡_int'] * 100
df['ì˜ì—…ì´ìµì§ì ‘'] = round(df['ì˜ì—…ì´ìµì§ì ‘'], 2)
df
```

```python
#18/19ë…„ë„ì— ëŒ€í•´ì„œ Price to Dream Ratio (PDR)ì„ ê³„ì‚°í•˜ì„¸ìš”. ë‹¨ ì—¬ê¸°ì„œ ì ìœ ìœ¨ì€ ì¼ë°˜ë‹´ë°° ê¸°ì¤€ì´ë©° 18ë…„:62%, 19ë…„ 63.5% ë¡œ ê³„ì‚°í•©ë‹ˆë‹¤. 
# ì‹œì¥ì˜ ê·œëª¨ëŠ” 18ë…„ 15ì¡°, 19ë…„ 16ì¡° ë¼ê³  ê°€ì •í•©ë‹ˆë‹¤.
# ì‹œê°€ì´ì•¡ì„ TAMì— í•´ë‹¹ ê¸°ì—…ì˜ ì‹œì¥ì ìœ ìœ¨ì„ ê³±í•œ ê°’ìœ¼ë¡œ ë‚˜ëˆ  PDRì„ ì‚°ì¶œí•œë‹¤. 
# ì˜ˆë¥¼ ë“¤ì–´ ì‹œê°€ì´ì•¡ 5ì¡°ì› ê·œëª¨ì˜ Aì—…ì²´ê°€ 100ì¡°ì›ì˜ ì‹œì¥ ê·œëª¨ë¥¼ ê°€ì§„ ì‚°ì—… ë‚´ì—ì„œ 10%ì˜ ì‹œì¥ì ìœ ìœ¨ì„ ê°–ê³  ìˆë‹¤ë©´ PDRì€ 0.5ë‹¤.

# 2018ë…„ : 13.8ì¡°(ì‹œì´)/15ì¡°*0.62
# 2019ë…„ : 12ì¡°/16ì¡°*0.635
```

```python
#2018ë…„  138,000 / (150,000*0.62)
#2019ë…„  120,000/160,000*0.635

#ë³€ìˆ˜ ê° ë…„ë„ë³„ ì‹œê°€ì´ì•¡
mc_2018 = 138000
mc_2019 = 120000
#TAM ë³€ìˆ˜
tam_2018 = 150000
tam_2019 = 160000
# ì‹œì¥ì ìœ ìœ¨ ë³€ìˆ˜
ms_2018 = 0.62
ms_2019 = 0.635

```

```python
pdr_dict = {'ì—°ë„':['2018','2019']
            ,'ì‹œê°€ì´ì•¡':[mc_2018, mc_2019]
            ,'TAM':[tam_2018,tam_2019]
            ,'ì‹œì¥ì ìœ ìœ¨':[ms_2018,ms_2019]
            }
```

```python
#ë°ì´í„°í”„ë ˆì„ë§Œë“¤ê¸°
df_pdr = pd.DataFrame(pdr_dict)
df_pdr.head()
df_pdr['PDR'] = df_pdr['ì‹œê°€ì´ì•¡']/(df_pdr['TAM']*df_pdr['ì‹œì¥ì ìœ ìœ¨'])
#######################1.2ë²ˆë‹µ#######################
# pdr ì¶”ê°€í•´ì„œ dataframe ì¶œë ¥
df_pdr.head()
```
